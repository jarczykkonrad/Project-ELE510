```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating image with different levels of brightness and contrast

img = np.ones((200, 400), dtype=np.uint8) * 255

font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, 'Clear Text', (10, 50), font, 1, (0), 3, cv2.LINE_AA)
cv2.putText(img, 'Faint Text', (10, 100), font, 1, (100), 3, cv2.LINE_AA)
cv2.putText(img, 'Dark Text', (10, 150), font, 1, (200), 3, cv2.LINE_AA)

# adding a gradient to the background
for i in range(img.shape[1]):
    img[:, i] = img[:, i] * (i / img.shape[1])

plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
# Global thresholding

global_thresh_value ,global_thresh_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(global_thresh_img, cmap='gray')
plt.axis('off')
```

```python
# Adaptive thresholding
adaptive_thresh_img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11 ,1)

plt.imshow(adaptive_thresh_img, cmap='gray')
plt.axis('off')
```
