```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

# Load the image in grayscale mode
img = cv.imread('./shadowed_object_1.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "File could not be read, check with os.path.exists()"

# Apply Gaussian filtering
blur = cv.GaussianBlur(img, (5, 5), 0)

# Apply Otsu's thresholding on the original image
ret1, th1 = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

# Apply Otsu's thresholding on the blurred image
ret2, th2 = cv.threshold(blur, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

# Set up the plots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
plt.subplots_adjust(hspace=0.5)

# Original image and histogram
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original Noisy Image')
axes[0, 0].axis('off')

axes[0, 1].hist(img.ravel(), 256, [0, 256])
axes[0, 1].set_title('Histogram for Original Image')

axes[0, 2].imshow(th1, cmap='gray')
axes[0, 2].set_title("Otsu's Thresholding")
axes[0, 2].axis('off')

# Blurred image and histogram
axes[1, 0].imshow(blur, cmap='gray')
axes[1, 0].set_title('Gaussian Filtered Image')
axes[1, 0].axis('off')

axes[1, 1].hist(blur.ravel(), 256, [0, 256])
axes[1, 1].set_title('Histogram for Blurred Image')

axes[1, 2].imshow(th2, cmap='gray')
axes[1, 2].set_title("Otsu's Thresholding on Blurred Image")
axes[1, 2].axis('off')

# Show the figure with all subplots
plt.show()
```
